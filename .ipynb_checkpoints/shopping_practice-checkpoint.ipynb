```python
import pandas as pd
```

```python
file_path = "Resources/shopping_data.csv"
shopping_df = pd.read_csv(file_path, encoding= "ISO-8859-1")
shopping_df.head(5)
```

```python
# columns
shopping_df.columns
```

```python
#find types
shopping_df.dtypes
```

```python
for column in shopping_df.columns:
    print(f"Column{column} has {shopping_df[column].isnull().sum()}null values")
```

```python
# Drop null rows
shopping_df =shopping_df.dropna()
shopping_df
```

```python
# find duplicate entries
print(f"Duplicate entries: {shopping_df.duplicated().sum()}")
```

```python
shopping_df.drop(columns=["CustomerID"], inplace = True)
shopping_df.head()
```

```python
# transfrom string column
def change_string(member):
    if member== "Yes":
        return 1
    else:
        return 0
shopping_df["Card Member"] = shopping_df["Card Member"].apply(change_string)
shopping_df.head()
```

```python
# transform annual income
shopping_df["Annual Income"] = shopping_df["Annual Income"] / 1000
shopping_df.head()
```

```python
# rename columns
shopping_df = shopping_df.rename(columns={'Card Member': 'card_member', 'Age': 'age', 'Annual Income' : 'annual_income', 'Spending Score (1-100)':'spending_score'})
# Or rename the existing DataFrame (rather than creating a copy) 
shopping_df.rename(columns={'Card Number': 'card_number', 'Age': 'age', 'Annual Income' : 'annual_income', 'Spending Score (1-100)':'spending_score'}, inplace=True)
shopping_df.head()
```

```python
# Saving cleaned data
file_path = "Resources/shopping_data_cleaned.csv"
shopping_df.to_csv(file_path, index=False)
```
